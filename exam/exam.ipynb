```
def all_start_with_capital(words):
    for w in words:
        if not w or not w[0].isupper():
            return False
        return True

test1 = all_start_with_capital(["hello", "world"])
print(test1)
test2 = all_start_with_capital(["Hello", "world"])
print(test2)

# you can see that after running, test 1 comes back false, because the string doesn't start with a capital, and vise versa
```
